```python
def checkSlope(df, col, window=5, stride=1):
    ser = df[col].values
    slope = []
    for i in range(ser.shape[0]):
        ys = ser[i:i+window]
#         print(f"ys shape: {ys.shape}")
        xs = np.arange(0, len(ys), 1)
#         print(f"xs shape: {xs.shape}")
        if ys.shape[0]>=5:
            trend = np.polyfit(x=xs, y=ys, deg=1)
            trend = np.around(trend[0], 3)
            slope.append(trend)
        else :
            slope.append(np.nan)
    df['slope'] = slope
    return df
```

```python
import pyageng
from scipy import signal
import matplotlib.pyplot as plt
# import dautil as dl
x = df_s['close'][100:200].dropna().values
```

```python
f, Pxx_den = signal.periodogram(x,fs=2, return_onesided=True)
plt.semilogy(f[1:], np.abs(Pxx_den[1:]))
plt.xlabel('Frequency (Hz)')
plt.plot(f[1:][signal.argrelmax(Pxx_den[1:])], np.abs(Pxx_den[1:][signal.argrelmax(Pxx_den[1:])]), 'o', label='Tops')
plt.plot(f[1:][signal.argrelmin(Pxx_den[1:])], np.abs(Pxx_den[1:][signal.argrelmin(Pxx_den[1:])]), 'o', label='Tops')
plt.ylabel('PSD')
plt.show()
```
